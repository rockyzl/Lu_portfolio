```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_stock_statistics(symbol):
    url = f"https://finance.yahoo.com/quote/{symbol}/key-statistics"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    statistics_table = soup.find_all("table", class_="table-qsp-stats Mt(10px)")
    if not statistics_table:
        print("No statistics table found for the given symbol.")
        return None

    rows = statistics_table[0].find_all("tr")

    # Extract the data from the table
    data = {}
    for row in rows:
        cells = row.find_all("td")
        if len(cells) == 2:
            data[cells[0].text] = cells[1].text

    return data

# Example usage
symbol = "AAPL"  # Replace with the symbol of the stock you want to retrieve data for
stock_data = get_stock_statistics(symbol)
if stock_data:
    df = pd.DataFrame.from_dict(stock_data, orient="index", columns=["Value"])
    print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_stock_statistics(symbol):
    url = f"https://finance.yahoo.com/quote/{symbol}/key-statistics"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    summary_section = soup.find("section", attrs={"data-test": "qsp-summary-table"})
    if not summary_section:
        print("No statistics found for the given symbol.")
        return None

    statistics_table = summary_section.find_all("table")
    if not statistics_table:
        print("No statistics table found for the given symbol.")
        return None

    rows = statistics_table[0].find_all("tr")

    # Extract the data from the table
    data = {}
    for row in rows:
        cells = row.find_all("td")
        if len(cells) == 2:
            data[cells[0].text] = cells[1].text

    return data

# Example usage
symbol = "AAPL"  # Replace with the symbol of the stock you want to retrieve data for
stock_data = get_stock_statistics(symbol)
if stock_data:
    df = pd.DataFrame.from_dict(stock_data, orient="index", columns=["Value"])
    print(df)
```

```python
import pandas as pd
import yfinance as yf

# Define a function to fetch stock data
def get_stock_info(ticker):
    stock = yf.Ticker(ticker)
    info = stock.info

    statistics = {
        'pegRatio': info.get('pegRatio', None),
        'marketCap': info.get('marketCap', None),
        'trailingPE': info.get('trailingPE', None),
        'forwardPE': info.get('forwardPE', None),
        'priceToSales': info.get('priceToSalesTrailing12Months', None)
    }

    return statistics

# Download the list of NASDAQ symbols
nasdaq_tickers = pd.read_csv('https://pkgstore.datahub.io/core/nasdaq-listings/nasdaq-listed_csv/data/7665719fb51081ba0bd834fde71ce822/nasdaq-listed_csv.csv')

# We'll only use a subset for this example to avoid making too many requests
nasdaq_tickers = nasdaq_tickers

# Create an empty DataFrame
df = pd.DataFrame()

for ticker in nasdaq_tickers['Symbol']:
    try:
        stats = get_stock_info(ticker)
        df = pd.concat([df, pd.DataFrame(stats, index=[ticker])])
    except:
        print(f"Failed to retrieve data for {ticker}")

# Save the DataFrame to a CSV file
df.to_csv('stock_statistics.csv')

print("Data saved to stock_statistics.csv")
```
